```python
import pandas as pd
df_telco = pd.read_csv('telco-customer-churn.csv')
df_telco.head()
```

```python
def difference(data):
    return data.max() - data.min()

q1 = df_telco.groupby(['InternetService','gender'])
q1['MonthlyCharges'].agg([('minimum','min'),('maximum','max'),('average','mean'),('difference (max-min)',difference)])
```

```python
df_telco.groupby(['SeniorCitizen','OnlineSecurity']).size()
```

```python
q3 = df_telco.groupby(['Churn','InternetService'])
q3['tenure'].agg([('Average','mean'),('Max-Min',difference)])
```
