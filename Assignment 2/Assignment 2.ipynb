# Numpy

TASKS 1: Read the “Attendance_RAW.txt” and save the data into a NumPy array.

```python
import numpy as np
path = 'Attendance_RAW.txt'

#Save the attendance as int
lines = [int(x.rstrip()) for x in open(path)]

#Save the data into a NumPy array
attendance = np.array(lines)
```

TASK 2: Shape the NumPy array to a structure where row represents the employees, column represents the working days.

```python
#row - employee; column: working days
present = attendance.reshape(14,15).T
present
```

TASK 3: 3.Construct a function that can calculate and return the list of average of attendance for each day.

```python
#calcualte and return list of average of attendance
def calcAve(attendance):
    return attendance.mean(axis=0)

aveAttendance = calcAve(present)

for x in range(len(aveAttendance)):
    print('Day',x,': ',aveAttendance[x])
```

Given a data file, “emp_detail.csv” which consists of the employee’s details: name and department. 

TASK 1: Using Pandas library, read the CSV data file in Python into a variable emp_detail.

```python
import pandas as pd

emp_detail = pd.read_csv('emp_detail.csv')
emp_detail
```

TASK 3: Using the NumPy object of the attendance record created earlier:

a.Convert the NumPy object into Data Frame, name it attendance.


```python
attendance= pd.DataFrame(present)
attendance
```


b.Merge the Data Frames emp_detail and attendance into a new data frame, name it emp_attendance. 

```python
#concat along the columns (axis=1)
emp_attendance = pd.concat([emp_detail, attendance], axis=1)
emp_attendance
```

Add a column with name “total” into the emp_attendance Data Frame that consists of the total no. of days the employee attended the event.

```python
emp_attendance['total'] = present.sum(axis=1)
emp_attendance
```

TASK 5: Add a column with name “rate” into the emp_attendance Data Frame that consists of the attendance rate / percentage the employee throughout the 14 days.

```python
emp_attendance['rate'] = emp_attendance['total']/len(attendance.columns)*100

emp_attendance
```

TASK 6: Consider following rules in classify the attendance:

a.Rate > 90% : Good

b.60% < Rate < 90% : Moderate

c.Rate < 60% : Bad

Based on the grouping rule above, get the status for each of the staff and save the status in a new column “status” in the Data Frame.

```python
list_status=[]

for x in range(len(emp_attendance)):
    
    if emp_attendance['rate'][x]>90:
        list_status.append('Good')

    elif emp_attendance['rate'][x]<60:
        list_status.append('Bad')

    else:
        list_status.append('Moderate')

emp_attendance['status']=list_status
emp_attendance
```

TASK 7: Construct a function that returns the average attendance of all staff over the 14 days according to the department. 

```python
def calcAverageAttendance(emp_attendance):
    total_a=0
    total_b=0
    
    for x in range(len(emp_attendance)):
        if emp_attendance['Department'][x]=='FINANCE':
            total_a+=emp_attendance['total'][x]
            Average_a = total_a / len(emp_attendance[emp_attendance['Department']=='FINANCE'])
    
        else:
            total_b+=emp_attendance['total'][x]
            Average_b = total_b / len(emp_attendance[emp_attendance['Department']=='STAFFING'])

                                
    return Average_a, Average_b                             
```

```python
Finance, Staffing = calcAverageAttendance(emp_attendance)

print('The average attendance of all staff over the 14 days:')
print('FINANCE: ', Finance)
print('STAFFING: ', Staffing)
```
